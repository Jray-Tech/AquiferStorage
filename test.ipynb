```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Generating random values
np.random.seed(0)  # For reproducibility
x = np.random.rand(100)
y = np.random.rand(100)
z = np.random.rand(100)

# Putting them in a DataFrame
df = pd.read_excel("data/dataset_1.xlsx")

df.head()
```

```python
col_to_plot  = [
    'Thickness of Aquifer (m) = TWL+TFL', 'Depth of Aquifer (m) = DWL',
       'AQUIFER RES of Weathered. \n(Ohm-M)',
       'Logarithm of W. Aquifer RESISTIVITY', 
      'Longitudinal Conductance (mhos)',
       'Transverse Resistance (Ωm²)', 'Transverse Resistivity (Ωm)',
       'Aquifer thickness TWA+TFA', 'fracture contrast',
       'aquifer storage in litres (i.e. *1000)'
]
```

```python
# Assuming col_to_plot and df are already defined
n_plots = len(col_to_plot)

# Define the maximum number of plots per row
max_plots_per_row = 3

# Calculate nrows and ncols
ncols = min(n_plots, max_plots_per_row)
nrows = (n_plots + max_plots_per_row - 1) // max_plots_per_row  # Ensures we have enough rows for all plots

# Adjust figsize: Width is adjusted based on ncols, and height is adjusted based on nrows
fig, axs = plt.subplots(nrows=nrows, ncols=ncols, figsize=(ncols*6, nrows*6), squeeze=False)  # Use squeeze=False for consistent indexing

for i, col in enumerate(col_to_plot):
    # Calculate row and column index
    row = i // ncols
    col_idx = i % ncols

    ax = axs[row, col_idx]
    contour = ax.tricontourf(df['Easting (Min)'], df['Northing (Min)'], df[col])
    fig.colorbar(contour, ax=ax, label=col)
    ax.set_title(f'Plot for {col}')
    ax.set_xlabel('Longitude (Easting)')
    ax.set_ylabel('Latitude (Northing)')

# Adjust the layout to make sure there's no overlap
plt.tight_layout()
plt.show()
```
